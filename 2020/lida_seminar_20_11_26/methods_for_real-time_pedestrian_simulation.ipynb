```python
# Imports
from IPython.display import HTML
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import seaborn as sns

%matplotlib inline
plt.style.use(['fivethirtyeight', 'dark_background'])
```

```python
def make_example_data():
    n = 50

    mean_trump_prior = 100
    sd_trump_prior = 1

    mean_observations = 50
    sd_observations = 10

    mean_posterior = (mean_observations * sd_trump_prior**2 + mean_trump_prior * sd_observations**2)/(sd_trump_prior**2 + sd_observations**2)
    sd_posterior = (sd_trump_prior**2 * sd_observations**2)/(sd_trump_prior**2 + sd_observations**2)
    
    trump_prior = np.random.normal(mean_trump_prior, sd_trump_prior, n)
    observations = np.random.normal(mean_observations, sd_observations, n)
    posterior = np.random.normal(mean_posterior, sd_posterior, n)
    
    return trump_prior, observations, posterior


def make_trump_graph(trump_prior):
    plt.figure(figsize=(12, 12))
    sns.kdeplot(trump_prior, fill=True, label='Trump prior', color='orange')
    plt.legend(loc='upper left')
    plt.xlabel('% Trump businesses that have been successful')
    plt.show()

def make_obs_graph(observations):
    plt.figure(figsize=(12, 12))
    sns.kdeplot(observations, fill=True, label='Observations', color='deepskyblue')
    plt.legend(loc='upper left')
    plt.xlabel('% Trump businesses that have been successful')
    plt.show()

def make_posterior_graph(posterior):
    plt.figure(figsize=(12, 12))
    sns.kdeplot(posterior, fill=True, label='Posterior', color='lime')
    plt.legend(loc='upper left')
    plt.xlabel('% Trump businesses that have been successful')
    plt.show()


def make_trump_obs_graph(trump_prior, observations):
    plt.figure(figsize=(16, 12))
    sns.kdeplot(trump_prior, fill=True, label='Trump prior', color='orange')
    sns.kdeplot(observations, fill=True, label='Observations', color='deepskyblue')
    plt.legend(loc='upper left')
    plt.xlabel('% Trump businesses that have been successful')
    plt.show()


def make_all_graph(trump_prior, observations, posterior):
    plt.figure(figsize=(16, 12))
    sns.kdeplot(trump_prior, fill=True, label='Trump prior', color='orange')
    sns.kdeplot(observations, fill=True, label='Observations', color='deepskyblue')
    sns.kdeplot(posterior, fill=True, label='Posterior', color='lime')
    plt.legend(loc='upper left')
    plt.xlabel('% Trump businesses that have been successful')
    plt.show()
```

```python
def make_1d_data(sd_prior=1, sd_obs=1):
    n = 100
    mean_prior = 10
    mean_obs = 20

    mean_posterior = (mean_obs * sd_prior**2 + mean_prior * sd_obs**2)/(sd_prior**2 + sd_obs**2)
    sd_posterior = (sd_prior**2 * sd_obs**2)/(sd_prior**2 + sd_obs**2)
    
    prior = np.random.normal(mean_prior, sd_prior, n)
    observations = np.random.normal(mean_obs, sd_obs, n)
    posterior = np.random.normal(mean_posterior, sd_posterior, n)
    
    return prior, observations, posterior


def make_1d_graph(prior, obs, posterior=None):
    plt.figure(figsize=(16, 12))
    sns.kdeplot(prior, fill=True, label='Prior', color='orange')
    sns.kdeplot(obs, fill=True, label='Observations', color='deepskyblue')
    if posterior is not None:
        sns.kdeplot(posterior, fill=True, label='Posterior', color='lime')
    plt.axvline(15, label='Truth', linestyle='--')
    plt.legend(loc='upper left')
    plt.xlabel('Pedestrian x-location')
    plt.show()
```

```python
def make_2d_data(sd_prior_x=1, sd_prior_y=1, sd_obs_x=1, sd_obs_y=1):
    n = 100
    mean_prior_x = 10
    mean_prior_y = 10
    mean_obs_x = 20
    mean_obs_y = 20

    mean_posterior_x = (mean_obs_x * sd_prior_x**2 + mean_prior_x * sd_obs_x**2)/(sd_prior_x**2 + sd_obs_x**2)
    sd_posterior_x = (sd_prior_x**2 * sd_obs_x**2)/(sd_prior_x**2 + sd_obs_x**2)
    mean_posterior_y = (mean_obs_y * sd_prior_y**2 + mean_prior_y * sd_obs_y**2)/(sd_prior_y**2 + sd_obs_y**2)
    sd_posterior_y = (sd_prior_y**2 * sd_obs_y**2)/(sd_prior_y**2 + sd_obs_y**2)
    
    prior_x = np.random.normal(mean_prior_x, sd_prior_x, n)
    observations_x = np.random.normal(mean_obs_x, sd_obs_x, n)
    posterior_x = np.random.normal(mean_posterior_x, sd_posterior_x, n)
    prior_y = np.random.normal(mean_prior_y, sd_prior_y, n)
    observations_y = np.random.normal(mean_obs_y, sd_obs_y, n)
    posterior_y = np.random.normal(mean_posterior_y, sd_posterior_y, n)
    
    return prior_x, prior_y, observations_x, observations_y, posterior_x, posterior_y


def make_2d_graph(prior_x, prior_y, obs_x, obs_y, posterior_x, posterior_y):
    plt.figure(figsize=(16, 12))
    sns.kdeplot(x=prior_x, y=prior_y, fill=True, label='Prior', color='orange')
    sns.kdeplot(x=obs_x, y=obs_y, fill=True, label='Observations', color='deepskyblue')
    sns.kdeplot(x=posterior_x, y=posterior_y, fill=True, label='Posterior', color='lime')
    plt.plot(15, 15, 'bo', markersize=20, label='Truth')

    plt.legend(loc='upper left')
    plt.xlabel('Pedestrian x-location')
    plt.ylabel('Pedestrian y-location')
    plt.show()
```

```python
# low uncertainty obs
prior_x_a, prior_y_a, obs_x_a, obs_y_a, posterior_x_a, posterior_y_a = make_2d_data(sd_prior_x=2, sd_prior_y=2)

# low uncertainty prior
prior_x_b, prior_y_b, obs_x_b, obs_y_b, posterior_x_b, posterior_y_b = make_2d_data(sd_obs_x=2, sd_obs_y=2)
```

# Solution: Data Assimilation

* The proposed solution to this problem is to use something called data assimilation
* Data assimilation is the term that we use to describe a collection of methods that allow us to combine/update the information that we have encapsulated in our model wtih observations.
* These methods are typically used in numerical weather prediction, where they have models of how a weather system will evolve over time, they collect observations of the system state, e.g. temperature measurements at a collection of locaations, and they use these observations to update the model state.
* In our case, we would like to apply the same approach, using agent-based models of pedestrian data alongside footfall data.

* A collection of methods used to combine information encapsulated in a model with observations.

* Frequently used in Numerical Weather Prediction

# Data Assimilation and Bayesian Updating

* As I mentioned before, thinking about data assimilation as a bayesian updating process wasn't something that was obvious to me when I first started my PhD, but the more I read around the subject the more it became apparent to me - data assimilation mirrors how a lot of us approach develop our beliefs about the world around us.
* We start off with some notion of how something works, we may then observe some new evidence regarding the phenomenon and consequently we go on to update how belief about it.
* Ultimately, this is what bayesian updating is - taking a prior belief, gathering observation evidence an updating that belief.

# Terminology

* Before proceeding, I just want to very quickly cover some terminology.
* You may have noticed that I used the word prior, and this was deliberate.
* When we talk about bayesian updating, we frequently refer to priors and posteriors.
* Priors are our belief about something before the evidence has been considered.
* Posteriors are our belief about something after the evidence has been considered.

* Prior: our belief about something *before* evidence is considered

* Posterior: our belief about something *after* evidence is considered

# Data Assimilation for Agent-Based Modelling

* You might then be thinking "this is great Keiran, but what does it have to do with real-time pedestrian simulation and agent-based modelling?"
* Well we can apply the same approach - the same line of thinking - about updating a model of pedestrianss with observations
* In this case, the prior is the model state before observations are considered
* And the posterior is the model state after the observations have been considered
* In this scenario, our aim is to more accurately estimate the positions of agents
* We measure accuracy using the difference between the position of a pedestrian estimated by the model and the true position of the pedestrian.

* Prior: model state *before* observations are considered

* Posterior: model state *after* observations are considered

* Reducing the error in the estimate of the agent's position

* Measure error by taking distance between estimated agent position and real agent position

# A Single Pedestrian in 1D

* Let us start by considering a single agent moving in 1-d - i.e. along the x-axis
* In this case, we again have our prior (which is where the model says that pedestrian is) and our observation.
* On the figure, I have denote the "true" location of the pedestrian with a dotted white line

```python
make_1d_graph(prior_1d_a, obs_1d_a)
```

# A Single Pedestrian in 1D

* Based on our prior and our observations, we can come up with a posterior, that is to say that we can update the model state based on the new evidence presented
* You may notice that, in this case, there is a larger degree of uncertainty in our prior model than in our observations - this is important: the updateed model posterior doesn't just depend on the locations given by the prior and the observations, but also on the difference in uncertainty.
* In this case, given that the observations have a lower uncertainty than the prior model state, the update to the model state errs more towards the observations.

```python
make_1d_graph(prior_1d_a, obs_1d_a, posterior_1d_a)
```

# A Single Pedestrian in 1D

* Conversely, if the uncertainty in our observations is comparatively higher than the uncertainty in our prior model state, then the update to the model state will err more towards the prior model state.

```python
make_1d_graph(prior_1d_b, obs_1d_b, posterior_1d_b)
```

# A Single Pedestrian in 2D

* We can extend this idea to modelling a single pedestrian in 2-d
* Once again, we characterise the prior model belief about the agent location with our orange distribution, the observation about the agent's location with the blue distribution and the updated posterior agent location in green.
* Again, we can see that, given that the distribution is more spread out, the uncertainty in the prior model estimate of the pedestrian location is more uncertain that the observed location, and so the updated posterior location leans more towards the observed location.

```python
make_2d_graph(prior_x_a, prior_y_a, obs_x_a, obs_y_a, posterior_x_a, posterior_y_a)
```

# A Single Pedestrian in 2D

* Similarly, if the there is greater uncertainty in the observed pedestrian location than in the prior location estimated by the model, the posterior location will lean more towards the prior model state.

```python
make_2d_graph(prior_x_b, prior_y_b, obs_x_b, obs_y_b, posterior_x_b, posterior_y_b)
```
