```python
# Imports
from IPython.display import HTML
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import seaborn as sns

%matplotlib inline
plt.style.use(['fivethirtyeight', 'dark_background'])
```

```python
def make_example_data():
    n = 50

    mean_trump_prior = 100
    sd_trump_prior = 1

    mean_observations = 50
    sd_observations = 10

    mean_posterior = (mean_observations * sd_trump_prior**2 + mean_trump_prior * sd_observations**2)/(sd_trump_prior**2 + sd_observations**2)
    sd_posterior = (sd_trump_prior**2 * sd_observations**2)/(sd_trump_prior**2 + sd_observations**2)
    
    trump_prior = np.random.normal(mean_trump_prior, sd_trump_prior, n)
    observations = np.random.normal(mean_observations, sd_observations, n)
    posterior = np.random.normal(mean_posterior, sd_posterior, n)
    
    return trump_prior, observations, posterior


def make_trump_graph(trump_prior):
    plt.figure(figsize=(12, 12))
    sns.kdeplot(trump_prior, fill=True, label='Trump prior', color='orange')
    plt.legend(loc='upper left')
    plt.xlabel('% Trump businesses that have been successful')
    plt.show()

def make_obs_graph(observations):
    plt.figure(figsize=(12, 12))
    sns.kdeplot(observations, fill=True, label='Observations', color='deepskyblue')
    plt.legend(loc='upper left')
    plt.xlabel('% Trump businesses that have been successful')
    plt.show()

def make_posterior_graph(posterior):
    plt.figure(figsize=(12, 12))
    sns.kdeplot(posterior, fill=True, label='Posterior', color='lime')
    plt.legend(loc='upper left')
    plt.xlabel('% Trump businesses that have been successful')
    plt.show()


def make_trump_obs_graph(trump_prior, observations):
    plt.figure(figsize=(16, 12))
    sns.kdeplot(trump_prior, fill=True, label='Trump prior', color='orange')
    sns.kdeplot(observations, fill=True, label='Observations', color='deepskyblue')
    plt.legend(loc='upper left')
    plt.xlabel('% Trump businesses that have been successful')
    plt.show()


def make_all_graph(trump_prior, observations, posterior):
    plt.figure(figsize=(16, 12))
    sns.kdeplot(trump_prior, fill=True, label='Trump prior', color='orange')
    sns.kdeplot(observations, fill=True, label='Observations', color='deepskyblue')
    sns.kdeplot(posterior, fill=True, label='Posterior', color='lime')
    plt.legend(loc='upper left')
    plt.xlabel('% Trump businesses that have been successful')
    plt.show()
```

```python
def make_1d_data(sd_prior=1, sd_obs=1):
    n = 100
    mean_prior = 10
    mean_obs = 20

    mean_posterior = (mean_obs * sd_prior**2 + mean_prior * sd_obs**2)/(sd_prior**2 + sd_obs**2)
    sd_posterior = (sd_prior**2 * sd_obs**2)/(sd_prior**2 + sd_obs**2)
    
    prior = np.random.normal(mean_prior, sd_prior, n)
    observations = np.random.normal(mean_obs, sd_obs, n)
    posterior = np.random.normal(mean_posterior, sd_posterior, n)
    
    return prior, observations, posterior


def make_1d_graph(prior, obs, posterior=None):
    plt.figure(figsize=(16, 12))
    sns.kdeplot(prior, fill=True, label='Prior', color='orange')
    sns.kdeplot(obs, fill=True, label='Observations', color='deepskyblue')
    if posterior is not None:
        sns.kdeplot(posterior, fill=True, label='Posterior', color='lime')
    plt.axvline(15, label='Truth', linestyle='--')
    plt.legend(loc='upper left')
    plt.xlabel('Pedestrian x-location')
    plt.show()
```

```python
def make_2d_data(sd_prior_x=1, sd_prior_y=1, sd_obs_x=1, sd_obs_y=1):
    n = 100
    mean_prior_x = 10
    mean_prior_y = 10
    mean_obs_x = 20
    mean_obs_y = 20

    mean_posterior_x = (mean_obs_x * sd_prior_x**2 + mean_prior_x * sd_obs_x**2)/(sd_prior_x**2 + sd_obs_x**2)
    sd_posterior_x = (sd_prior_x**2 * sd_obs_x**2)/(sd_prior_x**2 + sd_obs_x**2)
    mean_posterior_y = (mean_obs_y * sd_prior_y**2 + mean_prior_y * sd_obs_y**2)/(sd_prior_y**2 + sd_obs_y**2)
    sd_posterior_y = (sd_prior_y**2 * sd_obs_y**2)/(sd_prior_y**2 + sd_obs_y**2)
    
    prior_x = np.random.normal(mean_prior_x, sd_prior_x, n)
    observations_x = np.random.normal(mean_obs_x, sd_obs_x, n)
    posterior_x = np.random.normal(mean_posterior_x, sd_posterior_x, n)
    prior_y = np.random.normal(mean_prior_y, sd_prior_y, n)
    observations_y = np.random.normal(mean_obs_y, sd_obs_y, n)
    posterior_y = np.random.normal(mean_posterior_y, sd_posterior_y, n)
    
    return prior_x, prior_y, observations_x, observations_y, posterior_x, posterior_y


def make_2d_graph(prior_x, prior_y, obs_x, obs_y, posterior_x, posterior_y):
    plt.figure(figsize=(16, 12))
    sns.kdeplot(x=prior_x, y=prior_y, fill=True, label='Prior', color='orange')
    sns.kdeplot(x=obs_x, y=obs_y, fill=True, label='Observations', color='deepskyblue')
    sns.kdeplot(x=posterior_x, y=posterior_y, fill=True, label='Posterior', color='lime')
    plt.plot(15, 15, 'bo', markersize=20, label='Truth')

    plt.legend(loc='upper left')
    plt.xlabel('Pedestrian x-location')
    plt.ylabel('Pedestrian y-location')
    plt.show()
```
